```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Superstore Sales Dataset.csv')
```

```python
df
```

```python
order_counts = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_counts.columns = ['Customer ID', 'Order Frequency']
df = df.merge(order_counts, on='Customer ID', how='left')
```

```python
q_low = df['Sales'].quantile(0.33)
q_high = df['Sales'].quantile(0.66)
```

```python
def categorize_sales(sale):
    if sale <= q_low:
        return 'Low'
    elif sale <= q_high:
        return 'Medium'
    else:
        return 'High'

df['Sales Category'] = df['Sales'].apply(categorize_sales)
```
